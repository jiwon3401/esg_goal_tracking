```python
import os
import time
import shutil
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.chrome.options import Options
import chromedriver_autoinstaller

options = webdriver.ChromeOptions()
options.add_experimental_option('excludeSwitches', ['enable-logging'])
chromedriver_autoinstaller.install()
browser = webdriver.Chrome(options=options)

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

path_download = "./"
path = './Report_crawler/Report_set'
createFolder(path)

df = pd.read_excel("./Report_crawler/Report_list.xlsx")

for year, name, url in zip(df['year'], df['name'], df['url']) :
    folder_list = os.listdir(path)
    try : 
        browser.get(url)
        time.sleep(3)
        while True :
            filelist = os.listdir(path_download)
            filename = filelist.pop()
            if "crdownload" in filename :
                time.sleep(3)
            elif "tmp" in filename :
                time.sleep(3)
            else :
                break
                     
        if name in folder_list :
            shutil.move(path_download + '/' + filename, path + '/' + name + '/' + str(year) + '_' + name + '.pdf')
        else :
            createFolder(path + '/' + name)
            shutil.move(path_download + '/' + filename, path + '/' + name + '/' + str(year) + '_' + name + '.pdf ')
    except Exception as e:
        print(e)
        print(year + "_" + name, url)
        try :
            da = Alert(browser)
            da.accept()
            da.dismiss()
        except : 
            continue
        continue
```
